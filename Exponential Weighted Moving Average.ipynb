# Exponential Weighted Moving Average

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('DailyDelhiClimateTrain.csv')
df.head()
```

```python
df = df[['date', 'meantemp']]
```

```python
df.head()
```

```python
df.shape
```

```python
plt.scatter(df['date'], df['meantemp'], color='red')
plt.show()
```

```python
def ewma(beta, values):
    ewma_values = [values[0]]
    for i in range(1, len(values)):
        ewma = beta*ewma_values[i-1] + (1 - beta)*values[i]
        ewma_values.append(ewma)
    return ewma_values
```

```python
plt.title('beta : 0.1')
plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], ewma(beta=0.1, values=df['meantemp']), color='blue')
plt.show()
```

```python
plt.title('beta : 0.5')
plt.scatter(df['date'], df['meantemp'], color='red')
plt.plot(df['date'], ewma(beta=0.5, values=df['meantemp']), color='blue')
plt.show()
```
